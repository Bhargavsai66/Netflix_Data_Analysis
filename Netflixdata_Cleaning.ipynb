Importing necessary Libraries

```python
import pandas as pd
```

```python
df = pd.read_csv("netflix_titles.csv")
df
```

Data Cleaning

```python
df.info()
```

Dealing With Null Values

```python
df.isnull().sum()
```

```python
df['director'].unique() 
```

```python
df['director'].fillna('unknown', inplace = True)
df['cast'].fillna('unknown', inplace = True)
df['country'].fillna('unknown', inplace = True)
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset = ['date_added'])
```

```python
df['rating'].value_counts()
```

```python
df['rating'].fillna('TV-MA', inplace = True) 
```

```python
df = df.dropna(subset=['duration'])
```

```python
df.isnull().sum()
```

Dealing with duplicates records

```python
df.duplicated().sum()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
df['date_added'].unique()
```

```python
# 1. Remove leading/trailing spaces
df['date_added'] = df['date_added'].str.strip()

# 2. Convert to datetime safely
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df['duration_value'] = df['duration'].str.extract('(\d+)').astype(float)
df['duration_unit'] = df['duration'].str.extract('([A-Za-z]+)')
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df
```

```python
def convert(value):
    val = value[1:]
    return int(val)
```

```python
df['show_id'] = df['show_id'].apply(convert)
```

Final Cleaned_Dataset

```python
df
```
